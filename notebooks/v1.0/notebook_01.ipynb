# Justificación

Al evaluar la idea de emigrar quise basar mi desición en datos despues de descartar paises de mi continente (America). 

Al examinar Europa, me enfoque en aquellos paises con grandes tasas de crecimiento, pues si un pais es un tren, avanzas mas en un tren que va mas rapido. tomo el crecimiento como la medida central de mi evaluacion.

La medida mas conocida de crecimiento es el PIB (Producto Interior Bruto) comparativo en porcentaje, pero no es la mas acertada. Entonces decidi tomar otras medidas comparativas para tener una foto mas amplia de lo que ocurria en Europa antes de tomar una decisión. Spoiler... PL.

Este analisis se basará en informacion oficial de la API de eurostat organizada por indicadores clave relevantes:

    1. `notebook_02.ipynb`: Indice de convergencia.
    2. `notebook_03.ipynb`: Convergencia beta.
    3. `notebook_04.ipynb`: Convergencia sigma.
    4. `notebook_05.ipynb`: Comparación visual y proyecciones.
    5. `notebook_06.ipynb`: Conclusiones hallazgos y mejoras.

En la carpeta `docs/` está el diseño metodológico del estudio.
